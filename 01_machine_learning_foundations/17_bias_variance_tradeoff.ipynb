# Bias-Variance Tradeoff

In this notebook, we visualize underfitting, overfitting, bias, and variance using polynomial regression models of different complexities.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)

y = np.array([1,2,3,5,4,6,7,8])
```

```python
X_plot = np.linspace(1,8,100).reshape(-1,1)
```

```python
linear_model = LinearRegression()

linear_model.fit(X, y)

y_linear = linear_model.predict(X_plot)
```

```python
plt.scatter(X, y)

plt.plot(X_plot,
         y_linear,
         color='red')

plt.title("High Bias (Underfitting)")

plt.show()
```

```python
poly2 = PolynomialFeatures(degree=2)

X_poly2 = poly2.fit_transform(X)

X_plot_poly2 = poly2.transform(X_plot)

model2 = LinearRegression()

model2.fit(X_poly2, y)

y_poly2 = model2.predict(X_plot_poly2)
```

```python
plt.scatter(X, y)

plt.plot(X_plot,
         y_poly2,
         color='green')

plt.title("Balanced Bias-Variance")

plt.show()
```

```python
poly7 = PolynomialFeatures(degree=7)

X_poly7 = poly7.fit_transform(X)

X_plot_poly7 = poly7.transform(X_plot)

model7 = LinearRegression()

model7.fit(X_poly7, y)

y_poly7 = model7.predict(X_plot_poly7)
```

```python
plt.scatter(X, y)

plt.plot(X_plot,
         y_poly7,
         color='orange')

plt.title("High Variance (Overfitting)")

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.scatter(X, y)

plt.plot(X_plot,
         y_linear,
         label="High Bias",
         color='red')

plt.plot(X_plot,
         y_poly2,
         label="Balanced",
         color='green')

plt.plot(X_plot,
         y_poly7,
         label="High Variance",
         color='orange')

plt.legend()

plt.title("Bias-Variance Tradeoff")

plt.show()
```
