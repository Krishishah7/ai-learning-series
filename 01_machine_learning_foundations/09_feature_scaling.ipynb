# Feature Scaling

In this notebook, we explore normalization and standardization techniques used to scale features for Machine Learning models.

```python
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "Age": [18, 25, 35, 45, 55],
    "Salary": [20000, 35000, 60000, 80000, 120000]
}

df = pd.DataFrame(data)

df
```

```python
minmax_scaler = MinMaxScaler()

normalized_data = minmax_scaler.fit_transform(df)

normalized_df = pd.DataFrame(
    normalized_data,
    columns=df.columns
)

normalized_df
```

```python

standard_scaler = StandardScaler()

standardized_data = standard_scaler.fit_transform(df)

standardized_df = pd.DataFrame(
    standardized_data,
    columns=df.columns
)

standardized_df
```

```python
print("Original Data")
print(df)

print("\nNormalized Data")
print(normalized_df)

print("\nStandardized Data")
print(standardized_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Original
plt.subplot(1,3,1)
plt.scatter(df["Age"], df["Salary"])
plt.title("Original Data")

# Normalized
plt.subplot(1,3,2)
plt.scatter(normalized_df["Age"], normalized_df["Salary"])
plt.title("Normalized Data")

# Standardized
plt.subplot(1,3,3)
plt.scatter(standardized_df["Age"], standardized_df["Salary"])
plt.title("Standardized Data")

plt.show()
```
