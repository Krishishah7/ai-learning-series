# Hierarchical Clustering

In this notebook, we explore Hierarchical Clustering and visualize cluster formation using a dendrogram.

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
```

```python
X, y = make_blobs(
    n_samples=100,
    centers=4,
    random_state=42
)

print(X.shape)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.title(
    "Dataset"
)

plt.show()
```

```python
linked = linkage(
    X,
    method='ward'
)

plt.figure(figsize=(10,5))

dendrogram(linked)

plt.title(
    "Dendrogram"
)

plt.xlabel(
    "Data Points"
)

plt.ylabel(
    "Distance"
)

plt.show()
```

```python
model = AgglomerativeClustering(
    n_clusters=4
)

clusters = model.fit_predict(X)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.title(
    "Hierarchical Clustering"
)

plt.show()
```

```python
import pandas as pd

pd.Series(
    clusters
).value_counts()
```

## Conclusion

Hierarchical Clustering grouped data by progressively merging nearby points and clusters.

Observations:
- A dendrogram visualizes how clusters are formed.
- Agglomerative clustering starts with individual points and merges them step by step.
- The hierarchy provides insight into cluster relationships.
- Hierarchical clustering is useful when exploring data structure.
