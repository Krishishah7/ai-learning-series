# Cost Function in Linear Regression

In this notebook, we understand how to measure error in a model using a cost function.

The goal of a model is to minimize this error.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
m = 1.5
b = 1

y_pred = m * X + b
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')

for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='gray', linestyle='--')

plt.title("Error Visualization")
plt.show()
```

```python
errors = y - y_pred
```

```python
mse = np.mean(errors**2)
print("Mean Squared Error:", mse)
```

```python
m2 = 2
b2 = 0

y_pred2 = m2 * X + b2
mse2 = np.mean((y - y_pred2)**2)

print("Better MSE:", mse2)
```
