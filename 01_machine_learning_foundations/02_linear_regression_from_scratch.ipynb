# Linear Regression From Scratch

In this notebook, we implement Linear Regression without using any ML library.

We manually compute the best-fit line using mathematical formulas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
plt.scatter(X, y)
plt.title("Dataset")
plt.show()
```

```python
x_mean = np.mean(X)
y_mean = np.mean(y)
```

```python
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)

m = numerator / denominator
```

```python
b = y_mean - m * x_mean
```

```python
y_pred = m * X + b
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title("Regression Line (From Scratch)")
plt.show()
```

```python
print("Slope (m):", m)
print("Intercept (b):", b)
```

### How the model works

- The model tries to find a line: y = mx + b  
- m = slope → how steep the line is  
- b = intercept → where line crosses y-axis  

We calculate m using covariance and variance:
- Numerator → relationship between X and y  
- Denominator → spread of X  

This gives the best-fitting line.
