# Kernel SVM for Non-Linear Classification

In this notebook, we implement Kernel SVM using the RBF kernel to classify non-linearly separable data.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
from sklearn.svm import SVC
```

```python
X, y = make_circles(
    n_samples=200,
    noise=0.1,
    factor=0.4,
    random_state=42
)
```

```python
plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    s=50
)

plt.title("Non-Linear Dataset")

plt.show()
```

```python
linear_model = SVC(kernel='linear')

linear_model.fit(X, y)
```

```python
rbf_model = SVC(kernel='rbf')

rbf_model.fit(X, y)
```

```python
kernel='rbf'
```

```python
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.02),
    np.arange(y_min, y_max, 0.02)
)
```

```python
Z_linear = linear_model.predict(
    np.c_[xx.ravel(), yy.ravel()]
)

Z_linear = Z_linear.reshape(xx.shape)

plt.contourf(xx, yy, Z_linear, alpha=0.3)

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    s=50
)

plt.title("Linear SVM Boundary")

plt.show()
```

```python
Z_rbf = rbf_model.predict(
    np.c_[xx.ravel(), yy.ravel()]
)

Z_rbf = Z_rbf.reshape(xx.shape)

plt.contourf(xx, yy, Z_rbf, alpha=0.3)

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    s=50
)

plt.title("Kernel SVM (RBF) Boundary")

plt.show()
```

```python
print(
    "Linear SVM Accuracy:",
    linear_model.score(X, y)
)

print(
    "RBF Kernel SVM Accuracy:",
    rbf_model.score(X, y)
)
```
