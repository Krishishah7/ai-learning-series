```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons

from sklearn.cluster import (
    KMeans,
    AgglomerativeClustering,
    DBSCAN
)
```

```python
X, y = make_moons(
    n_samples=300,
    noise=0.08,
    random_state=42
)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.title(
    "Original Dataset"
)

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=2,
    random_state=42
)

kmeans_labels = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    X[:,0],
    X[:,1],
    c=kmeans_labels
)

plt.title(
    "K-Means"
)

plt.show()
```

```python
hierarchical = AgglomerativeClustering(
    n_clusters=2
)

hierarchical_labels = hierarchical.fit_predict(X)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    X[:,0],
    X[:,1],
    c=hierarchical_labels
)

plt.title(
    "Hierarchical Clustering"
)

plt.show()
```

```python
dbscan = DBSCAN(
    eps=0.2,
    min_samples=5
)

dbscan_labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    X[:,0],
    X[:,1],
    c=dbscan_labels
)

plt.title(
    "DBSCAN"
)

plt.show()
```

```python
import pandas as pd

comparison = pd.DataFrame({
    "Algorithm":[
        "K-Means",
        "Hierarchical",
        "DBSCAN"
    ],

    "Requires K":[
        "Yes",
        "Yes",
        "No"
    ],

    "Handles Outliers":[
        "No",
        "Limited",
        "Yes"
    ]
})

comparison
```

## Conclusion

This notebook compared three clustering algorithms.

Observations:

- K-Means works well for compact spherical clusters.
- Hierarchical Clustering builds a cluster hierarchy.
- DBSCAN handles irregular shapes and outliers effectively.
- Different datasets may require different clustering approaches.
