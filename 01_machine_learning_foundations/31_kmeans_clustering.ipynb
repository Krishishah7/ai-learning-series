# K-Means Clustering

In this notebook, we use K-Means clustering to discover hidden groups in unlabeled data.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(
    n_samples=300,
    centers=3,
    random_state=42
)

print(X.shape)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.title(
    "Unlabeled Data"
)

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)
```

```python
kmeans.fit(X)
```

```python
clusters = kmeans.labels_

print(clusters[:20])
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.title(
    "K-Means Clusters"
)

plt.show()
```

```python
print(
    kmeans.cluster_centers_
)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=300,
    marker='*',
    color="red"
)

plt.title(
    "Clusters with Centroids"
)

plt.show()
```

```python
df = pd.DataFrame({
    "Cluster": clusters
})

df["Cluster"].value_counts()
```

## Conclusion

K-Means clustering grouped unlabeled data into three clusters.

Observations:
- Similar data points were grouped together.
- Cluster centers (centroids) represented the center of each cluster.
- K-Means is an unsupervised learning algorithm because it does not require labels.
