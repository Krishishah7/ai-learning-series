# DBSCAN Clustering

In this notebook, we use DBSCAN to discover clusters based on density and identify outliers automatically.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X, y = make_moons(
    n_samples=300,
    noise=0.08,
    random_state=42
)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.title(
    "Dataset"
)

plt.show()
```

```python
dbscan = DBSCAN(
    eps=0.2,
    min_samples=5
)

clusters = dbscan.fit_predict(X)
```

```python
print(
    pd.Series(clusters).value_counts()
)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.title(
    "DBSCAN Clustering"
)

plt.show()
```

```python
n_clusters = len(
    set(clusters)
) - (1 if -1 in clusters else 0)

print(
    "Number of Clusters:",
    n_clusters
)
```

```python
outliers = list(clusters).count(-1)

print(
    "Outliers:",
    outliers
)
```

## K-Means vs DBSCAN

K-Means:
- Requires K
- Sensitive to outliers
- Works best on spherical clusters

DBSCAN:
- Does not require K
- Detects outliers automatically
- Handles irregular cluster shapes

## Conclusion

DBSCAN grouped data based on density and automatically identified outliers.

Observations:
- Clusters are formed where data points are densely packed.
- Outliers are labeled as -1.
- DBSCAN can discover irregular cluster shapes.
- It is useful when the number of clusters is unknown.
