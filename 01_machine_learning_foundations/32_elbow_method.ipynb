# Elbow Method for K-Means

In this notebook, we use the Elbow Method to determine the optimal number of clusters for K-Means clustering.

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(
    n_samples=300,
    centers=4,
    random_state=42
)

print(X.shape)
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.title("Dataset")

plt.show()
```

```python
wcss = []
for k in range(1, 11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42
    )

    kmeans.fit(X)

    wcss.append(
        kmeans.inertia_
    )
```

```python
for k, value in enumerate(
    wcss,
    start=1
):
    print(
        f"K={k}, WCSS={value:.2f}"
    )
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker='o'
)

plt.xlabel(
    "Number of Clusters (K)"
)

plt.ylabel(
    "WCSS"
)

plt.title(
    "Elbow Method"
)

plt.grid()

plt.show()
```

## Conclusion

The Elbow Method was used to determine the optimal number of clusters for K-Means.

Observations:
- WCSS decreases as K increases.
- The elbow point indicates a suitable number of clusters.
- Choosing K using the Elbow Method helps avoid under-clustering and over-clustering.
