# Gradient Descent in Linear Regression

In this notebook, we implement Gradient Descent to minimize error and find the best-fit line automatically.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
m = 0
b = 0

learning_rate = 0.01
epochs = 100
n = len(X)
```

```python
losses = []
```

```python
for i in range(epochs):

    y_pred = m * X + b

    # calculate mse
    mse = np.mean((y - y_pred) ** 2)
    losses.append(mse)

    # gradients
    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db
```

```python
y_pred = m * X + b
```

```python
print("Initial Error:", losses[0])
print("Final Error:", losses[-1])
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.title("Error Reduction Over Time")
plt.show()
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.title("Gradient Descent Result")
plt.show()
```

```python
print("Final slope:", m)
print("Final intercept:", b)
```
