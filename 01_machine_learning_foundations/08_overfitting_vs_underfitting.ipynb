# Overfitting vs Underfitting

In this notebook, we explore how model complexity affects learning and generalization.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)

y = np.array([1,3,2,5,7,8,8,9])
```

```python
linear_model = LinearRegression()

linear_model.fit(X, y)

y_linear = linear_model.predict(X)
```

```python
plt.scatter(X, y)

plt.plot(X, y_linear, color='red')

plt.title("Underfitting")

plt.show()
```

```python
poly2 = PolynomialFeatures(degree=2)

X_poly2 = poly2.fit_transform(X)

model2 = LinearRegression()

model2.fit(X_poly2, y)

y_poly2 = model2.predict(X_poly2)
```

```python
plt.scatter(X, y)

plt.plot(X, y_poly2, color='green')

plt.title("Good Fit")

plt.show()
```

```python
poly7 = PolynomialFeatures(degree=7)

X_poly7 = poly7.fit_transform(X)

model7 = LinearRegression()

model7.fit(X_poly7, y)

y_poly7 = model7.predict(X_poly7)
```

```python
plt.scatter(X, y)

plt.plot(X, y_poly7, color='orange')

plt.title("Overfitting")

plt.show()
```
