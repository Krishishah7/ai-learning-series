# Linear Regression Basics

Linear Regression is one of the simplest and most important algorithms in Machine Learning.  
It is used to model the relationship between an input variable (X) and an output variable (y) using a straight line.

The goal is to find the best-fitting line that predicts y from X.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])
```

```python
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dataset")
plt.show()
```

The data shows a clear linear relationship between X and y.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title("Linear Regression Line")
plt.show()
```

```python
print("Slope (Coefficient):", model.coef_)
print("Intercept:", model.intercept_)
```

Slope tells how much y changes when X increases.  
Intercept is the value of y when X = 0.
