# Epochs and Batches

Neural networks usually do not process the entire training dataset at once.

Instead, the dataset is divided into smaller batches. The model processes one batch at a time, updates its weights, and repeats until all batches are processed.

One complete pass through the entire training dataset is called an **epoch**.

Training usually consists of multiple epochs, allowing the network to gradually improve its predictions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.arange(1, 21)

print("Dataset:")
print(data)
```

```python
batch_size = 5

batches = []

for i in range(0, len(data), batch_size):
    batch = data[i:i+batch_size]
    batches.append(batch)

for index, batch in enumerate(batches, start=1):
    print(f"Batch {index}: {batch}")
```

```python
num_batches = len(batches)

print("Total Batches:", num_batches)
```

```python
epochs = 3

for epoch in range(1, epochs + 1):
    print(f"\nEpoch {epoch}")

    for index, batch in enumerate(batches, start=1):
        print(f" Processing Batch {index}: {batch}")
```

```python
iterations = epochs * num_batches

print("Epochs:", epochs)
print("Batches per Epoch:", num_batches)
print("Total Iterations:", iterations)
```

```python
comparison = pd.DataFrame({
    "Batch Size": [1, 5, 10, 20],
    "Batches per Epoch": [
        20,
        4,
        2,
        1
    ]
})

comparison
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    comparison["Batch Size"].astype(str),
    comparison["Batches per Epoch"]
)

plt.xlabel("Batch Size")
plt.ylabel("Number of Batches")
plt.title("Effect of Batch Size")

plt.show()
```

```python
steps = [
    "Dataset",
    "Batch",
    "Forward Pass",
    "Loss",
    "Backpropagation",
    "Weight Update",
    "Next Batch",
    "Next Epoch"
]

plt.figure(figsize=(12,2))

plt.plot(
    range(len(steps)),
    [1]*len(steps),
    marker="o"
)

for i, step in enumerate(steps):
    plt.text(i, 1.05, step, ha="center")

plt.xticks([])
plt.yticks([])

plt.title("Neural Network Training Process")

plt.show()
```
