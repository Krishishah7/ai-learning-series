# Perceptron Basics

A Perceptron is the simplest type of artificial neural network.

It was introduced by Frank Rosenblatt in 1958 and is designed for binary classification problems.

A perceptron receives input values, assigns weights to them, computes a weighted sum, adds a bias, applies an activation function, and produces an output.

The perceptron is considered the building block of modern deep learning models.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours_studied = 6
attendance = 8

print("Hours Studied:", hours_studied)
print("Attendance:", attendance)
```

```python
weight_hours = 0.7
weight_attendance = 0.5

print("Weight (Hours):", weight_hours)
print("Weight (Attendance):", weight_attendance)
```

```python
bias = -5

print("Bias:", bias)
```

```python
weighted_sum = (
    hours_studied * weight_hours +
    attendance * weight_attendance +
    bias
)

print("Weighted Sum:", weighted_sum)
```

```python
def step_function(x):
    if x >= 0:
        return 1
    return 0
```

```python
prediction = step_function(weighted_sum)

print("Prediction:", prediction)
```

```python
students = [

    [2,3],

    [6,8],

    [4,5],

    [8,9]

]

for student in students:

    weighted_sum = (
        student[0] * weight_hours +
        student[1] * weight_attendance +
        bias
    )

    prediction = step_function(weighted_sum)

    print(student, "Prediction:", prediction)
```

```python
x = np.linspace(-10,10,200)

y = []

for value in x:
    y.append(step_function(value))

plt.figure(figsize=(8,5))

plt.plot(x,y)

plt.title("Step Activation Function")

plt.xlabel("Input")

plt.ylabel("Output")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(6,6))

pass_students = np.array([
    [6,8],
    [7,9],
    [8,7]
])

fail_students = np.array([
    [2,3],
    [3,2],
    [4,3]
])

plt.scatter(
    pass_students[:,0],
    pass_students[:,1],
    label="Pass"
)

plt.scatter(
    fail_students[:,0],
    fail_students[:,1],
    label="Fail"
)

plt.xlabel("Hours Studied")

plt.ylabel("Attendance")

plt.title("Simple Binary Classification")

plt.legend()

plt.grid()

plt.show()
```
