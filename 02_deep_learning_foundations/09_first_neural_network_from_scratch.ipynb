# 09 - First Neural Network From Scratch
Build a simple neural network using NumPy.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset

```
X=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=float)
y=np.array([[0],[0],[0],[1]],dtype=float)
X,y
```

## Sigmoid

```
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_derivative(x):
    return x*(1-x)
```

## Initialize Parameters

```
np.random.seed(42)
weights=np.random.randn(2,1)
bias=np.zeros((1,1))
learning_rate=0.1
epochs=1000
loss_history=[]
```

## Train

```
for epoch in range(epochs):
    z=np.dot(X,weights)+bias
    pred=sigmoid(z)
    loss=-np.mean(y*np.log(pred+1e-8)+(1-y)*np.log(1-pred+1e-8))
    loss_history.append(loss)
    error=pred-y
    dW=np.dot(X.T,error)/len(X)
    dB=np.mean(error)
    weights-=learning_rate*dW
    bias-=learning_rate*dB
    if epoch%100==0:
        print(epoch,round(loss,4))
```

## Loss Curve

```
plt.figure(figsize=(8,5))
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs Epoch")
plt.grid()
plt.show()
```

## Predictions

```
pred=sigmoid(np.dot(X,weights)+bias)
binary=(pred>=0.5).astype(int)
results=pd.DataFrame({
'Input1':X[:,0].astype(int),
'Input2':X[:,1].astype(int),
'Actual':y.flatten().astype(int),
'Probability':np.round(pred.flatten(),4),
'Predicted':binary.flatten()
})
results
```

## Accuracy

```
accuracy=np.mean(binary==y)*100
print(f'Accuracy: {accuracy:.2f}%')
```

## Workflow

```
steps=['Input','Forward','Prediction','Loss','Backprop','Gradient Descent','Updated Weights']
plt.figure(figsize=(11,2))
plt.plot(range(len(steps)),[1]*len(steps),marker='o')
for i,s in enumerate(steps):
    plt.text(i,1.05,s,ha='center')
plt.xticks([]);plt.yticks([])
plt.title('Neural Network Training Workflow')
plt.show()
```
