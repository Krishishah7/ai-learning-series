# Gradient Descent

Gradient Descent is an optimization algorithm used to minimize the loss function during neural network training.

It works by repeatedly updating the model's weights in the direction that reduces the prediction error.

The process continues until the model reaches a point where the loss is minimized.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def loss_function(weight):
    return (weight - 3) ** 2
```

```python
def gradient(weight):
    return 2 * (weight - 3)
```

```python
weight = 8

learning_rate = 0.1

print("Initial Weight:", weight)
```

```python
weights = []
losses = []

for i in range(20):

    current_loss = loss_function(weight)

    weights.append(weight)

    losses.append(current_loss)

    grad = gradient(weight)

    weight = weight - learning_rate * grad
```

```python
results = pd.DataFrame({
    "Iteration": range(1, 21),
    "Weight": np.round(weights, 4),
    "Loss": np.round(losses, 4)
})

results
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    results["Iteration"],
    results["Weight"],
    marker="o"
)

plt.title("Weight Convergence")

plt.xlabel("Iteration")
plt.ylabel("Weight")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    results["Iteration"],
    results["Loss"],
    marker="o"
)

plt.title("Loss Reduction")

plt.xlabel("Iteration")
plt.ylabel("Loss")

plt.grid()

plt.show()
```

```python
learning_rates = [0.01, 0.1, 0.5]

comparison = pd.DataFrame({
    "Learning Rate": learning_rates,
    "Behavior": [
        "Slow Convergence",
        "Balanced Learning",
        "Fast Convergence (may overshoot in complex problems)"
    ]
})

comparison
```

```python
steps = [
    "Initialize Weight",
    "Forward Pass",
    "Compute Loss",
    "Compute Gradient",
    "Update Weight",
    "Repeat"
]

plt.figure(figsize=(10,2))

plt.plot(range(len(steps)), [1]*len(steps), marker="o")

for i, step in enumerate(steps):
    plt.text(i, 1.05, step, ha="center")

plt.xticks([])
plt.yticks([])

plt.title("Gradient Descent Workflow")

plt.show()
```
