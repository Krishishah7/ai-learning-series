# Loss Functions

A loss function measures how far a model's prediction is from the actual value.

The objective of a neural network is to minimize this loss during training.

Lower loss indicates better predictions, while higher loss indicates larger prediction errors.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
actual = np.array([10, 20, 30, 40])

predicted = np.array([12, 18, 33, 38])

print("Actual:", actual)
print("Predicted:", predicted)
```

```python
errors = actual - predicted

print(errors)
```

```python
squared_errors = errors ** 2

mse = np.mean(squared_errors)

print("Squared Errors:", squared_errors)

print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import mean_squared_error

print(
    "Sklearn MSE:",
    mean_squared_error(actual, predicted)
)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    range(len(errors)),
    squared_errors
)

plt.xlabel("Sample")

plt.ylabel("Squared Error")

plt.title("Squared Error for Each Prediction")

plt.show()
```

```python
actual = np.array([1, 0, 1, 1])

predicted_probability = np.array([
    0.95,
    0.20,
    0.80,
    0.60
])
```

```python
bce = -np.mean(

    actual*np.log(predicted_probability)

    +

    (1-actual)

    *

    np.log(1-predicted_probability)

)

print("Binary Cross Entropy:", round(bce,4))
```

```python
good_prediction = 0.95

bad_prediction = 0.55

print("Good Prediction Error:", -np.log(good_prediction))

print("Bad Prediction Error:", -np.log(bad_prediction))
```

```python
comparison = pd.DataFrame({

    "Loss Function":[

        "Mean Squared Error",

        "Binary Cross Entropy"

    ],

    "Used For":[

        "Regression",

        "Binary Classification"

    ]

})

comparison
```

```python
loss_values = [

    mse,

    bce

]

labels = [

    "MSE",

    "Binary Cross Entropy"

]

plt.figure(figsize=(6,4))

plt.bar(
    labels,
    loss_values
)

plt.ylabel("Loss")

plt.title("Example Loss Values")

plt.show()
```
