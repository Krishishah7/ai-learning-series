# Forward Propagation

Forward Propagation is the process of passing input data through a neural network to generate predictions.

Each neuron receives inputs, computes a weighted sum, applies an activation function, and passes the output to the next layer until the final prediction is produced.

It is called "forward" because information moves only from the input layer to the output layer.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([2, 3])

print("Input:", X)
```

```python
W1 = np.array([
    [0.5, 0.2],
    [0.3, 0.8]
])

b1 = np.array([0.1, -0.2])
```

```python
W2 = np.array([
    [0.7],
    [0.6]
])

b2 = np.array([0.3])
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
z1 = np.dot(X, W1) + b1

print("Hidden Layer Weighted Sum:")
print(z1)
```

```python
a1 = relu(z1)

print("Hidden Layer Output:")
print(a1)
```

```python
z2 = np.dot(a1, W2) + b2

print("Output Layer Weighted Sum:")
print(z2)
```

```python
prediction = sigmoid(z2)

print("Final Prediction:")
print(prediction)
```

```python
steps = [
    "Input",
    "Weighted Sum",
    "ReLU",
    "Weighted Sum",
    "Sigmoid",
    "Prediction"
]

plt.figure(figsize=(10,2))

plt.plot(range(len(steps)), [1]*len(steps), marker='o')

for i, step in enumerate(steps):
    plt.text(i, 1.05, step, ha='center')

plt.yticks([])
plt.xticks([])
plt.title("Forward Propagation Flow")

plt.show()
```

```python
summary = {
    "Layer": [
        "Input Layer",
        "Hidden Layer",
        "Output Layer"
    ],
    "Operation": [
        "Receive Input",
        "Weighted Sum + ReLU",
        "Weighted Sum + Sigmoid"
    ]
}

import pandas as pd

summary_df = pd.DataFrame(summary)

summary_df
```
