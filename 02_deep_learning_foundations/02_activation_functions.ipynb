# Activation Functions

Activation functions determine whether a neuron should be activated based on its weighted input.

Without activation functions, neural networks behave like simple linear models and cannot learn complex patterns.

Different activation functions are used for different types of deep learning problems.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 200)
```

```python
def step(x):
    return np.where(x >= 0, 1, 0)

step_output = step(x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

sigmoid_output = sigmoid(x)
```

```python
def tanh(x):
    return np.tanh(x)

tanh_output = tanh(x)
```

```python
def relu(x):
    return np.maximum(0, x)

relu_output = relu(x)
```

```python
def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)

leaky_output = leaky_relu(x)
```

```python
plt.figure(figsize=(7,5))

plt.plot(x, step_output)

plt.title("Step Activation Function")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(x, sigmoid_output)

plt.title("Sigmoid Activation Function")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(x, tanh_output)

plt.title("Tanh Activation Function")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(x, relu_output)

plt.title("ReLU Activation Function")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(x, leaky_output)

plt.title("Leaky ReLU Activation Function")

plt.grid()

plt.show()
```
