# Backpropagation

Backpropagation is the learning mechanism of a neural network.

After a prediction is made, the loss function measures the prediction error. Backpropagation sends this error backward through the network and computes how each weight contributed to the error. The weights are then updated to reduce future prediction errors.

This process is repeated many times during training until the model learns useful patterns.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
weight = 0.5

print("Initial Weight:", weight)
```

```python
learning_rate = 0.1

print("Learning Rate:", learning_rate)
```

```python
gradient = 0.8

print("Gradient:", gradient)
```

```python
new_weight = weight - learning_rate * gradient

print("Updated Weight:", new_weight)
```

```python
weights = [weight]

current_weight = weight

for i in range(10):

    current_weight = current_weight - learning_rate * gradient

    weights.append(current_weight)

weights
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(len(weights)),
    weights,
    marker="o"
)

plt.title("Weight Updates During Training")

plt.xlabel("Iteration")

plt.ylabel("Weight Value")

plt.grid()

plt.show()
```

```python
loss = [2.0, 1.6, 1.3, 1.0, 0.8, 0.6, 0.5]

plt.figure(figsize=(8,5))

plt.plot(
    range(len(loss)),
    loss,
    marker="o"
)

plt.title("Loss Decreasing During Training")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.grid()

plt.show()
```

```python
steps = [
    "Input",
    "Forward Pass",
    "Prediction",
    "Loss",
    "Backpropagation",
    "Weight Update"
]

plt.figure(figsize=(11,2))

plt.plot(range(len(steps)), [1]*len(steps), marker="o")

for i, step in enumerate(steps):
    plt.text(i, 1.05, step, ha="center")

plt.xticks([])
plt.yticks([])

plt.title("Neural Network Training Workflow")

plt.show()
```
