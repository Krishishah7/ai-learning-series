# Gradient Descent Variants

This notebook explores Batch Gradient Descent, Stochastic Gradient Descent (SGD), and Mini-Batch Gradient Descent and compares their behavior.

## Imports

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Create Example Dataset

```
data = np.arange(1, 21)
print("Dataset:", data)
```

## Batch Gradient Descent

```
print("Batch Gradient Descent")
print("- Uses the entire dataset before updating weights.")
```

## Stochastic Gradient Descent (SGD)

```
print("Stochastic Gradient Descent")
for sample in data[:5]:
    print(f"Updating weight using sample {sample}")
print("...")
```

## Mini-Batch Gradient Descent

```
batch_size = 5
batches = [data[i:i+batch_size] for i in range(0, len(data), batch_size)]

for idx, batch in enumerate(batches, start=1):
    print(f"Batch {idx}: {batch}")
```

## Comparison Table

```
comparison = pd.DataFrame({
    "Variant":[
        "Batch Gradient Descent",
        "Stochastic Gradient Descent",
        "Mini-Batch Gradient Descent"
    ],
    "Data Used per Update":[
        "Entire Dataset",
        "One Sample",
        "Small Batch"
    ],
    "Speed":[
        "Slow",
        "Fast",
        "Balanced"
    ],
    "Stability":[
        "High",
        "Low",
        "Medium"
    ]
})
comparison
```

## Visualization

```
labels = ["Batch","SGD","Mini-Batch"]
updates = [1,20,4]

plt.figure(figsize=(6,4))
plt.bar(labels, updates)
plt.xlabel("Gradient Descent Variant")
plt.ylabel("Weight Updates per Epoch")
plt.title("Comparison of Gradient Descent Variants")
plt.show()
```

## Advantages & Disadvantages

```
summary = pd.DataFrame({
    "Variant":["Batch","SGD","Mini-Batch"],
    "Advantage":[
        "Stable updates",
        "Fast learning",
        "Balanced speed and stability"
    ],
    "Disadvantage":[
        "Slow on large datasets",
        "Noisy updates",
        "Requires choosing batch size"
    ]
})
summary
```

## Key Takeaways

```
print("Key Takeaways")
print("- Batch GD updates once per epoch.")
print("- SGD updates after every sample.")
print("- Mini-Batch GD updates after every batch.")
print("- Mini-Batch Gradient Descent is the most commonly used approach in modern deep learning.")
```
